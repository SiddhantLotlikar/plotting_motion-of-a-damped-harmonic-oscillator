```python
#developing a funtion to return differential of x wrt time and differential of v wrt time
def dSdx(x, S):
    global m,c,k
    x, v = S
    return [v, -(c*v + k*x)/m]
x_0 = 0
v_0 = 1
S_0 = (x_0, v_0)
```

```python
#creating an array of time intervals
t = np.linspace(0, 100, 100)
sol = odeint(dSdx, y0=S_0, t=t, tfirst=True)
```

```python
x_sol = sol.T[0]
v_sol = sol.T[1]
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=t, y=x_sol, label='Displacement (x)')
sns.lineplot(x=t, y=v_sol, label='Velocity (v)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Displacement and Velocity Over Time')
plt.legend()
plt.show()
```
